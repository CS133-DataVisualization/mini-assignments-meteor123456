```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
# to make sure that the plot generated by seaborn can be rendered and displayed inline on the
                   # jupyter notebook.
import seaborn as sns

marvel_url="https://raw.githubusercontent.com/csbfx/advpy122-data/master/marvel-wikia-data.csv"
```

```python
marvel = pd.read_csv(marvel_url)
marvel.head()
```

```python
marvel['HAIR'].value_counts()
```

```python
hair_palette = {
    "Black Hair": "#000000",
    "Brown Hair": "#8B4513",
    "Blond Hair": "#FFD700",
    "No Hair": "#D2D2D2",   # Light gray for no hair
    "Bald": "#E5C29F",      # Skin-like color
    "White Hair": "#E3E3E3",
    "Red Hair": "#FF4500",
    "Grey Hair": "#808080",
    "Green Hair": "#00FF00",
    "Auburn Hair": "#A52A2A",
    "Blue Hair": "#1E90FF",
    "Strawberry Blond Hair": "#FFB347",
    "Purple Hair": "#800080",
    "Orange Hair": "#FFA500",
    "Variable Hair": "#BEBEBE",  # Neutral gray for variability
    "Pink Hair": "#FFC0CB",
    "Yellow Hair": "#FFFF00",
    "Silver Hair": "#C0C0C0",
    "Gold Hair": "#FFD700",
    "Reddish Blond Hair": "#FFCC99",
    "Light Brown Hair": "#D2B48C",
    "Magenta Hair": "#FF00FF",
    "Orange-brown Hair": "#CD853F",
    "Bronze Hair": "#CD7F32",
    "Dyed Hair": "#B22222"  # Dyed hair can be a distinct, strong color
}
life_palette = {
    "Living Characters": "#1E90FF",
    "Deceased Characters":"#BEBEBE"
}
```

```python
import warnings # To suppress some warnings

# Suppress the specific FutureWarning
warnings.filterwarnings("ignore", category=FutureWarning, module="seaborn")
```

#What is the distribution of characters appearances based on their character hair colors?


```python
sns.catplot(
    data=marvel[marvel['APPEARANCES']<500],
    x="APPEARANCES",
    y="HAIR",
    aspect=1, # aspect change the aspect ratio
    height=15,
   # jitter=0.7, # jitter spread out the data points
    # palette="Set2",
    hue="HAIR",
    palette=hair_palette
    )
plt.title("Distribution of characters with less than 500 Appearances based on hair color");
```

```python
sns.catplot(
    data=marvel,
    x="APPEARANCES",
    y="HAIR",
    kind='box',
    aspect=1, # aspect change the aspect ratio
    height=15,
   # jitter=0.7, # jitter spread out the data points
    # palette="Set2",
    hue="HAIR",
    showfliers=False,
    palette=hair_palette
    )
plt.title("Character Appearances (Excluding Outliers) based on Hair Color ");
```

```python
showfliers=Falseg=sns.catplot(
    data=marvel,
    x="ALIGN",
    y="Year",
    kind="violin",
    cut=0,
    hue="ALIVE",
    bw_method=0.5,
    density_norm="count",
    split=True,
    palette = life_palette
    )

plt.title("Distribution of Year of First Appearances of Characters based on their Alignment and their Life Status");
```

Most of the characters of all alignments debuted the most between the 1990's and 2000's. We can also see that there is significantly more bad characters debuting over time compared to good and neutral characters.
